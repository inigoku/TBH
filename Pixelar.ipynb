```python
from PIL import Image
from numpy import asarray
from os import listdir
from os.path import isfile, join

img=[]
img_weights=[]
img.append(asarray(Image.open("/content/drive/MyDrive/Colab Notebooks/Yerba.jpeg"))[::16, ::16])
img_weights.append(200)
img.append(asarray(Image.open("/content/drive/MyDrive/Colab Notebooks/Árbol.jpeg"))[::16, ::16])
img_weights.append(100)
img.append(asarray(Image.open("/content/drive/MyDrive/Colab Notebooks/Cíclope.jpg"))[::16, ::16])
img_weights.append(2)
img.append(asarray(Image.open("/content/drive/MyDrive/Colab Notebooks/Demonio.jpg")))
img_weights.append(1)
ruta = "/content/drive/MyDrive/Colab Notebooks/Javier/"
for arch in listdir(ruta):
  fullarch = join(ruta, arch)
  if isfile(fullarch):
    img.append(asarray(Image.open(fullarch)))
    img_weights.append(1)
num_sprites = len(img)
```

```python
import matplotlib.pyplot as plt
def plot_char(image_data):
#  image = image_data.reshape(8, 8)
  image = image_data
  plt.imshow(image, cmap="gray")
  plt.axis("off")
```

```python
import random
plot_char(img[random.randint(0, num_sprites-1)])
```

```python
# genera mapa
import random
map = random.choices(range(num_sprites), weights=img_weights, k=1600)
plt.figure(figsize=(9, 9))
for idx in range(1600):
  plt.subplot(40, 40, idx + 1)
  plot_char(img[map[idx]])
plt.subplots_adjust(wspace=0, hspace=0)
```
